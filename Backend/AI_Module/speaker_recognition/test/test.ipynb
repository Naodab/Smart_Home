```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def visualize_audio_features_with_mfcc(audio_path1, audio_path2, sr=16000, n_mfcc=13):
    # Load hai tp m thanh
    y1, _ = librosa.load(audio_path1, sr=sr)
    y2, _ = librosa.load(audio_path2, sr=sr)

    # Tnh MFCC
    mfcc1 = librosa.feature.mfcc(y=y1, sr=sr, n_mfcc=n_mfcc, hop_length=512, n_fft=2048)
    mfcc2 = librosa.feature.mfcc(y=y2, sr=sr, n_mfcc=n_mfcc, hop_length=512, n_fft=2048)

    # V biu 
    plt.figure(figsize=(12, 8))

    # Biu  sng m thanh 1
    plt.subplot(2, 2, 1)
    librosa.display.waveshow(y1, sr=sr)
    plt.title("Waveform - Audio 1")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    # Biu  MFCC m thanh 1
    plt.subplot(2, 2, 2)
    librosa.display.specshow(mfcc1, sr=sr, hop_length=512, x_axis='time', cmap='viridis')
    plt.title("MFCC - Audio 1")
    plt.colorbar(format='%+2.0f')

    # Biu  sng m thanh 2
    plt.subplot(2, 2, 3)
    librosa.display.waveshow(y2, sr=sr)
    plt.title("Waveform - Audio 2")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    # Biu  MFCC m thanh 2
    plt.subplot(2, 2, 4)
    librosa.display.specshow(mfcc2, sr=sr, hop_length=512, x_axis='time', cmap='viridis')
    plt.title("MFCC - Audio 2")
    plt.colorbar(format='%+2.0f')

    plt.tight_layout()
    plt.show()

# V d s dng
audio_path1 = "../data/Hoang/Hoang_01.wav"
audio_path2 = "../data/Hoang/angry.wav"
visualize_audio_features_with_mfcc(audio_path1, audio_path2)
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial.distance import cosine

def visualize_audio_similarity(audio_path1, audio_path2, sr=16000, n_mfcc=13):
  # Load hai tp m thanh
  y1, _ = librosa.load(audio_path1, sr=sr)
  y2, _ = librosa.load(audio_path2, sr=sr)

  # Tnh MFCC
  mfcc1 = librosa.feature.mfcc(y=y1, sr=sr, n_mfcc=n_mfcc, hop_length=512, n_fft=2048)
  mfcc2 = librosa.feature.mfcc(y=y2, sr=sr, n_mfcc=n_mfcc, hop_length=512, n_fft=2048)

  # Tnh khong cch cosine gia cc vector MFCC trung bnh
  mfcc1_mean = np.mean(mfcc1, axis=1)
  mfcc2_mean = np.mean(mfcc2, axis=1)
  cosine_distance = cosine(mfcc1_mean, mfcc2_mean)

  # V biu 
  plt.figure(figsize=(14, 10))

  # Biu  sng m thanh 1
  plt.subplot(3, 2, 1)
  librosa.display.waveshow(y1, sr=sr)
  plt.title("Waveform - Audio 1")
  plt.xlabel("Time (s)")
  plt.ylabel("Amplitude")

  # Biu  MFCC m thanh 1
  plt.subplot(3, 2, 2)
  librosa.display.specshow(mfcc1, sr=sr, hop_length=512, x_axis='time', cmap='viridis')
  plt.title("MFCC - Audio 1")
  plt.colorbar(format='%+2.0f')

  # Biu  sng m thanh 2
  plt.subplot(3, 2, 3)
  librosa.display.waveshow(y2, sr=sr)
  plt.title("Waveform - Audio 2")
  plt.xlabel("Time (s)")
  plt.ylabel("Amplitude")

  # Biu  MFCC m thanh 2
  plt.subplot(3, 2, 4)
  librosa.display.specshow(mfcc2, sr=sr, hop_length=512, x_axis='time', cmap='viridis')
  plt.title("MFCC - Audio 2")
  plt.colorbar(format='%+2.0f')

  # Biu  so snh MFCC trung bnh
  plt.subplot(3, 2, 5)
  plt.plot(mfcc1_mean, label="Audio 1", marker='o')
  plt.plot(mfcc2_mean, label="Audio 2", marker='x')
  plt.title("Mean MFCC Comparison")
  plt.xlabel("MFCC Coefficients")
  plt.ylabel("Value")
  plt.legend()

  # Hin th khong cch cosine
  plt.subplot(3, 2, 6)
  plt.text(0.5, 0.5, f"Cosine Distance: {cosine_distance:.4f}", fontsize=14, ha='center', va='center')
  plt.axis('off')

  plt.tight_layout()
  plt.show()

# V d s dng
audio_path1 = "../data/Hoang/Hoang_01.wav"  # m thanh ca ngi 1
audio_path2 = "../data/Huy/Huy_01.wav"      # m thanh ca ngi 2
visualize_audio_similarity(audio_path1, audio_path2)
```
